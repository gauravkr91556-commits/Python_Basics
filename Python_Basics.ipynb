Question 2: Explain the difference between List and Tuple in Python.

| List                     | Tuple                         |
| ------------------------ | ----------------------------- |
| Mutable (can be changed) | Immutable (cannot be changed) |
| Uses [] brackets         | Uses () brackets              |
| Slower                   | Faster                        |
| Supports modification    | Does not support modification |


```python
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print(my_list)
print(my_tuple)
```

Question 5: Write a Python program to check whether a number is even or odd using conditional statements.

```python
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(number, "is Even")
else:
    print(number, "is Odd")
```

Question 7: Create a list of five numbers and print the maximum number from the list.

```python
numbers = [15, 8, 45, 32, 20]

print("List:", numbers)
print("Maximum Number:", max(numbers))
```

  Question 8: Write a Python program to remove duplicate values from a list using a set.

```python
numbers = [2, 3, 4, 2, 5, 2]

unique_numbers = list(set(numbers))

print("Original List:", numbers)
print("List after Removing Duplicates:", unique_numbers)
```

Question 9: Write a function that returns the square of a number.

```python
def square(num):
    return num ** 2

number = int(input("Enter a number: "))

print("Square =", square(number))
```

Question 10: Write a Python program to count how many times a number appears in a list.

```python
numbers = [2, 3, 4, 2, 5, 2]

search = int(input("Enter the number to count: "))

count = numbers.count(search)

print("List:", numbers)
print(search, "appears", count, "times.")
```
